```python
from random import randint

class Account:
    accountCnt = 0
    
    def __init__(self, name, money):
        self.name = name
        self.money = money
        self.bankName = 'SC은행'
        self.creditNumber = self.__create_credit_number()
        self.moneyInCnt = 0
        Account.accountCnt += 1
        self.deposit_history = []
        self.withdraw_history = []
    
    def __create_credit_number(self):
        return (str(randint(100,999)).zfill(3) + '-' +
            str(randint(10,99)).zfill(2) + '-' +
            str(randint(100000,999999)).zfill(6))
    
    @classmethod
    def get_account_num(cls):
        return cls.accountCnt

    def deposit(self, inMoney):
        if inMoney > 0:
            self.money += inMoney
            self.moneyInCnt += 1
            if self.moneyInCnt == 5:
                self.__sum_interest()
                self.moneyInCnt == 0
            self.deposit_history.append(inMoney)
        else:
            print('0원이하는 입금 불가')
    
    def __sum_interest(self):
        self.money += self.money*0.01
    
    def withdraw(self, outMoney):
        if outMoney <= self.money:
            self.money -= outMoney
            self.withdraw_history.append(outMoney) 
        elif outMoney > self.money:
            print('잔액 부족')
        else:
            print('출금오류')
    
    def display_info(self):
        print('은행이름: ', self.bankName)
        print('예금주: ', self.name)
        print('계좌번호: ', self.creditNumber)
        print('잔고: %s원' % self.__money_display(self.money))
    
    def __money_display(self, money):
        count = 0
        result = []
        point_money = str(money).split('.')[1]
        for i in str(int(money))[::-1]:
            count += 1
            result.append(i)
            if count % 3 == 0:
                result.append(',')
        if result[-1] == ',':
            del result[-1]
        
        return ''.join(result[::-1]) + '.' + point_money
        
    def print_deposit_history(self):
        print('입금 내역')
        for history in self.deposit_history:
            print(history)
    
    def print_withdraw_history(self):
        print('출금 내역')
        for history in self.withdraw_history:
            print(history)
        

account1 = Account('박장수',10000)
account2 = Account('김철수',1000000000)
account3 = Account('유새석',23234458)
print(Account.get_account_num())

account3.deposit(1000)
account3.deposit(1000)
account3.deposit(1000)
account3.deposit(1000)
account3.deposit(1000)
account3.withdraw(23000000)


account3.display_info()
account3.print_deposit_history()
account3.print_withdraw_history()

                
        
    
        
```
