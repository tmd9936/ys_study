```python
# [문제] Person 클래스를 정의하시오.
# 속성 : name, age
# 메소드 : eat, sleep, work

# 위에서 정의한 Person 클래스를 상속받는
# 객체(Student, employee)를 만들어보세요.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self, minute):
        print(f'{self.name}은 {minute}분동안 밥을 먹습니다.')
    
    def sleep(self, minute):
        print(f'{self.name}은 {minute}분동안 잠을 잡니다.')
    
    def work(self, minute):
        print(f'{self.name}은 {minute}분동안 일을 합니다.')

# Student가 아르바이트를  할 경우
# 일도 하고 공부도 한다.
# work() 호출시 "홍길동은 30분동안 일을 합니다."
#               "그리고 30분동안 공부도 합니다."

class Student(Person):
    def __init__(self, name, age, stu_num, is_part_time):
        super().__init__(name,age)
        self.stu_num = stu_num
        self.is_part_time = is_part_time
    
    def work(self, minute):
        if self.is_part_time:
            super().work(minute)
            print(f'그리고 {minute}분동안 공부를 합니다.')
        else:
            print(f'{self.name}은 {minute}분동안 공부를 합니다.')
        
    
class Employee(Person):
    def __init__(self, name, age, company):
        super().__init__(name,age)
        self.company = company
        
    def work(self, minute):
        print(f'{self.name}은 {minute}분동안 사무를 봅니다.')

    def night_overtime(self, minute):
        print(f'{self.name}은 {minute}분동안 야근을 합니다.')

# __init_가 정의되어있지 않으면 상위 클래스의 __init__를 그대로 가져옴
class President(Person):
    pass
    
    
p1 = Person('아무개', 48)
p1.work(100)
p1.eat(30)
p1.sleep(400)
print()

s1 = Student('길동', 14, '1277426', False)
s1.work(300)
s1.sleep(400)
print()

s1 = Student('수성', 18, '1877462', True)
s1.work(30)
s1.sleep(400)
print()

e1 = Employee('찬수', 30, '삼성')
e1.work(200)
e1.night_overtime(100)

pre1 = President('김씨', 50)
pre1.work(400)

    
```

```python
class Phone:
    """ Phone class """
    def __init__(self, maker, details):
        self.maker = maker
        self.details = details
    
    def phone_info(self):
        return f'{self.maker} : {self.details}'

# 스마트폰 객체를 만드세요..  maker는 원하는 걸로
# details는 color, price
class Galaxy(Phone):
    def __init__(self, makers, color, price, cpu):
        details = {
            "color" : color,
            "price" : price,
            "cpu"   : cpu
        }
        super().__init__(makers, details)

class SmartPhone(Phone):
    pass


        
g1 = Galaxy('samsung','gold',800000,'cp200')
g1.phone_info()

g2 = SmartPhone('LG',{'color':'red','price':10000})
print(g2.details['color'])
        

```
