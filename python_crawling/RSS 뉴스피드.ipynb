```python
# RSS : 사이트에서 보내주는 소식지(새로운 소식 제공)

import urllib.request
import urllib.parse

rss_url = 'https://news.sbs.co.kr/news/SectionRssFeed.do'
# https://news.sbs.co.kr/news/SectionRssFeed.do?sectionId=07&plink=RSSREADER

params = []

for query in ['01','02','07']:
    params.append(dict(sectionId=query))

print(params)

for c in params:
    param =  urllib.parse.urlencode(c)
    print(param)
    
    # URL 만들기
    url = rss_url + '?' + param + '&plink=RSSREADER'
    
    print('url =', url)
    
    res_data = urllib.request.urlopen(url).read()
    
    # 응답 데이터를 디코딩
    contents = res_data.decode('utf-8')
    
    print('#'*100)
    print(contents)
```
