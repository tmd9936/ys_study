```python
import requests
from bs4 import BeautifulSoup
import csv
# import asyncio

def movie_review(start, end):
    result = []
    
    for i in range(start, end+1):
        print(f'=================={i}Page===============')
        url = f'https://movie.naver.com/movie/point/af/list.nhn?&page={i}'

        res = requests.get(url)
        bs = BeautifulSoup(res.text, 'lxml')

        trs = bs.select('table.list_netizen > tbody > tr')
        
        for tr in trs:
            tds = tr.select('td')

            if len(tds) != 3:
                continue

            num = tds[0].text
        #     print(num)

            title = tds[1].select('a')[0].text
            point = tds[1].select('div > em')[0].text
            author = tds[2].select('a')[0].string
            # text = tds[1].text.split('\n')[5]
            # text = tds[1].select_one('br').nextSibling.strip()
            
            # extract() 함수: 태그및 태그의 내용까지 모두 제거함
            [x.extract() for x in tds[1].select('a')]
            [x.extract() for x in tds[1].select('div')]
            text = tds[1].text.strip()

            print(title, point, author)
            print(text)
            print()
            
            result.append([num, title, point, author, text])
        print(f'=================={i}Page===============')
        print()
    return result


reviews = movie_review(1,5)

with open('./resource/movie_review.csv', 'w', newline='', encoding='utf-8') as fp:
    _write = csv.writer(fp)
    _write.writerows(reviews)
          

    
    
    

```
