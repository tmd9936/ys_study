```python
import random
import os

words_dict = {
    '파이썬' : 'python',
    '호랑이' : 'tiger',
    '고양이' : 'cat',
    '사과'   : 'apple',
    '사자'   : 'lion'
}

words = []

# words_dict에서 키값만 빼서 리스트에 저장
for word in words_dict:
    words.append(word)

# words에 있는 값들을 뒤섞기
random.shuffle(words)

# 각 단어당 3회의 기회
chance = 3

os.system('cls')

for i in range(0, len(words_dict)):
    
    question = words[i]
    
    for j in range(0, chance):
        answer = input(f'{question}의 영어 단어를 입력하세요 > \n')
        
        english = words_dict[question]
        
        if answer.strip().lower() == english.lower():
            print('정답!')
            break
        else:
            print('땡!')
    else:
        print(f'{question}의 정답은 {english} 입니다.')
        
print('문제가 모두 출제~')
```
