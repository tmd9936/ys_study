```python
from bs4 import BeautifulSoup

html = """
<!DOCTYPE html>
<html lang="en">
<head>
    <title>Document</title>
</head>
<body>
    <h1>제목1</h1>
    <h2>제목2</h2>
    <p class="title"><strong>옛날 동화</strong></p>
    <p class="story">
        아주 옛날 이야기 ....
        <a href="http://test.com" class='sister' id='link1'>말똥이</a>
        <a href="http://test.com" class='sister' id='link2'>개똥이</a>
        <a href="http://test.com" class='sister3' id='link3'>길동이</a>
        <a href="http://test.com" data2='brother' class='sister3' id='link3'>길동이</a>
    </p>
    <p class="story">이야기 끝</p>
</body>
</html>
"""

# find 는 태그로 접근
# select는 스타일시트 형태로 접근

# 단일 : find, select_one
# 복수 : find_all, select


soup = BeautifulSoup(html, 'html.parser')

# soup 타입 확인
print('soup 타입 : ', type(soup))

# 코드를 보기 좋게 정리
print('prettify : ', soup.prettify())
print()

# h1 태그 액세스
h1 = soup.html.body.h1

print("h1 태그 :", h1)

# p 태그 액세스
p1 = soup.html.body.p
print('p1 태그 :', p1)

p2 = p1.next_sibling.next_sibling
print('p2 태그 :', p2)

# 컨텐트(텍스트)를 출력
print('h1태그의 텍스트:', h1.string)
print('p1태그의 텍스트:', p1.string)

print(p2.next_element.next_element)
# print(list(p2.next_element.next_elements))

## find, find_all 사용

# a태그를 모두 선택하기
link1 = soup.find_all('a')
print('link1 type : ', type(link1))

#첫 번째 a태그만 선택
link2 = soup.find('a')
print(link2)

link3 = soup.find_all('a', class_='sister')
print(link3)

link4 = soup.find('a', class_="sister")
print(link4)

link5 = soup.find('a', {'class':'sister3', 'data2':'brother'})
print(link5)

# select(find_all), select_one(find)
link6 = soup.select_one('p.title > strong')
print(link6)

link7 = soup.select_one('a#link2')
print(link7)
print(link7.string)

link8 = soup.select('p.story > a')
print(link8)

# p.story 밑에 두번째 a태그 가져오기
link9 = soup.select_one('p.story > a:nth-of-type(2)')
print(link9)
```
