```python
import json
import requests

sess = requests.Session()

# httpbin.org 사이트에 10개의 json 데이터를 요청해보기
res = sess.get('http://httpbin.org/stream/10')

# 수신데이터 확인
# print(res.text)

# print(type(res))

# Encoding 확인하기
print(f'Encoding: {res.encoding}')


if res.encoding is None:
    res.encoding = 'UTF-8'
    
print(f'Encoding: {res.encoding}')

# 수신된 데이터를 dict 타입으로 변환
for line in res.iter_lines(decode_unicode=True):
    print(line)
    print(type(line))
    
    # 타입이 str -> dict으로 바뀜
    # Json(Dict) 변환 후 타입 확인하기
    b = json.loads(line)
    print(b)
    print(type(b))
    
    for keys, values in b.items():
        print(f'keys: {keys}, Values : {values}')
    print()
    print()
```
